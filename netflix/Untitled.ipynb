```python
X = netflix.drop('Close',axis=1)
y = netflix['Close']
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sclr = StandardScaler()
```

```python

X_train = sclr.fit_transform(X_train)
X_test = sclr.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
def pred(Open,High,Low,Adj_Close,Volume,year,month,day):
    features = np.array([[Open,High,Low,Adj_Close,Volume,year,month,day]])
    features = sclr.fit_transform(features)
    prediction = lr.predict(features).reshape(1,-1)
    return prediction[0]
```

```python
Open = 200.45
High = 250.45
Low = 150.45
Adj_Close = 100.34
Volume = 100.45
year = 2020
month = 8
day = 4
 
res = pred(Open,High,Low,Adj_Close,Volume,year,month,day)
```
